```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv("../data/1 - Weekday Schedule/Baseline NB-Table 1.csv")
```

```python
new_cols = ["km", "From:"]

for item in zip(df.columns[2:], df.iloc[0][2:]):
    new_cols.append(str(item[0] + "_" + str(item[1])))

df.columns = new_cols
```

```python
data = df.iloc[1:]
```

```python
data.head()
```

```python
for col in data.columns[2:]:
    data[col] = pd.to_timedelta(data[col]+":00", errors="coerce")
```

```python
data
```

```python
data.to_csv("../data/parsed_data/schedules/baseline_NB.csv", index=False)
```
